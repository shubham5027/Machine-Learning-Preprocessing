```python
def rename(name):
    if name == "Royal Challengers Bangalore":
        return "RCB"
    if name == "Delhi Daredevils" :
        return "DD"
    if name == "Delhi Daredevils" :
        return "DD"
    if name == "Kolkata Knight Riders":
        return "KKR"
    if name =="Mumbai Indians":
        return "MI"
    else:
        return name
rename("Royal Challengers Bangalore")
```

```python
df3=pd.read_csv('IPL Matches 2008-2020.csv',converters={'team1':rename})
```
