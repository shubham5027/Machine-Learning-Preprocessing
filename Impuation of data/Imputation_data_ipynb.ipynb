```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split

from sklearn.impute import SimpleImputer
from sklearn.impute import MissingIndicator
from sklearn.impute import KNNImputer  #Multivariate Analysis (adavaced)
from sklearn.experimental import enable_iterative_imputer  #Multivariate analysis (adavace
from sklearn.impute import IterativeImputer
```

```python
df=pd.read_csv("titanic_toy.csv")
```

```python
df
```

```python
X=df.drop(['Survived'],axis=1)
Y=df['Survived']
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,random_state=0)
X_train
```

```python
df.isnull().sum()
```

Simple Imputer with Statregy

```python
si=SimpleImputer(strategy='mean')
si2=SimpleImputer(strategy="median")
si3=SimpleImputer(strategy="most_frequent")
```

```python
x_train_im=si.fit_transform(X_train[['Age','Fare']])
x_test_im=si.transform(X_test[['Age','Fare']])
```

```python
x_train_im2=si2.fit_transform(X_train)
x_test_im2=si2.transform(X_test)
```

```python
x_train_im3=si3.fit_transform(X_train[['Age','Fare']])
x_test_im3=si3.transform(X_test[['Age','Fare']])
```

KNN Imputer

```python
knn=KNNImputer(n_neighbors=5,weights='distance')
knn=KNNImputer(n_neighbors=5,weights='uniform')
```

```python
x_train_knn=knn.fit_transform(X_train)
x_test_knn=knn.transform(X_test)
```

```python
itm=IterativeImputer()
```

```python
x_train_itm=itm.fit_transform(X_train)
x_test_itm=itm.transform(X_test)
```
